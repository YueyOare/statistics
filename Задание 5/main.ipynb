```python
from sklearn.preprocessing import OneHotEncoder

# Выбор категориальных переменных (предположим, что они находятся в столбцах 'категориальная_переменная')
categorical_cols = ['Пол', 'Образование', 'Трудовая деятельность', 'Семейное положение', 'Проживание', 'Жилищные условия', 'Инвалидность', 'Отягощ. Наследственность', 'Инфекционные болезни НС', 'Злоупотребление психоактивными вещ-вами', 'Интоксикации', 'ЧМТ', 'Гипертония', 'Церебрально-сосудистые нарушения', 'Эндокринные заболевания', 'Аллергические заб., в т.ч. астма', 'Совпадение начала заболевания', 'Левша']

# Применение OneHotEncoder
encoder = OneHotEncoder(sparse=False, drop='first')  # drop='first' для избежания dummy variable trap
encoded_data = pd.DataFrame(encoder.fit_transform(data[categorical_cols]), columns=encoder.get_feature_names_out(categorical_cols))

# Объединение закодированных данных с исходным DataFrame
data = pd.concat([data, encoded_data], axis=1)

# Удаление старых категориальных столбцов
data = data.drop(categorical_cols, axis=1)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Предположим, что у вас есть обученная модель multinom_model и тестовые данные X_test, y_test
y_pred = model.predict(X)

# Построение таблицы сопряженности
conf_matrix = confusion_matrix(y, y_pred)

# Визуализация таблицы сопряженности с использованием seaborn
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y_pred), yticklabels=np.unique(y))
plt.xlabel('Предсказанные значения')
plt.ylabel('Фактические значения')
plt.title('Таблица сопряженности')
plt.show()
```

```python
# Извлекаем значения для вычисления чувствительности и специфичности
TP = np.diag(conf_matrix)
FP = np.sum(conf_matrix, axis=0) - TP
FN = np.sum(conf_matrix, axis=1) - TP
TN = np.sum(conf_matrix) - (TP + FP + FN)

# Вычисляем чувствительность и специфичность для каждого класса
sensitivity = TP / (TP + FN)
specificity = TN / (TN + FP)

# Выводим значения чувствительности и специфичности для каждого класса
for i, class_label in enumerate(model.classes_):
    print(f'Класс {class_label}: Чувствительность = {sensitivity[i]}, Специфичность = {specificity[i]}')
```

```python
from sklearn.metrics import roc_curve, auc

def find_optimal_thresholds(model, X, y):
    """
    Находит оптимальные пороговые значения для многоклассовой модели.

    Параметры:
    - model: обученная многоклассовая модель
    - X: массив признаков
    - y: массив истинных классов

    Возвращает:
    - thresholds: словарь, содержащий оптимальные пороговые значения для каждого класса
    """
    y_probs = model.predict_proba(X)

    thresholds = {}
    for i in range(model.classes_.shape[0]):
        fpr, tpr, threshold = roc_curve((y == model.classes_[i]).astype(int), y_probs[:, i])
        optimal_idx = np.argmax(tpr - fpr)
        optimal_threshold = threshold[optimal_idx]
        thresholds[model.classes_[i]] = optimal_threshold

    return thresholds
```

```python
thresholds = find_optimal_thresholds(model, X, y)
dict(sorted(thresholds.items(), key=lambda item: item[1]))
```

```python
def predict_with_thresholds(model, X, thresholds):
    """
    Применяет пороговые значения для многоклассовой модели.

    Параметры:
    - model: обученная многоклассовая модель
    - X: массив признаков
    - thresholds: словарь с оптимальными пороговыми значениями для каждого класса

    Возвращает:
    - y_pred: массив предсказанных классов с учетом порогов
    """
    y_probs = model.predict_proba(X)
    y_pred = np.zeros_like(y_probs)

    for i in range(model.classes_.shape[0]):
        y_pred[:, i] = (y_probs[:, i] > thresholds[model.classes_[i]]).astype(int)

    return model.classes_[np.argmax(y_pred, axis=1)]
```

```python
y_pred = predict_with_thresholds(model, X, thresholds)

# Построение таблицы сопряженности
conf_matrix = confusion_matrix(y, y_pred)

# Визуализация таблицы сопряженности с использованием seaborn
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y_pred), yticklabels=np.unique(y_test))
plt.xlabel('Предсказанные значения')
plt.ylabel('Фактические значения')
plt.title('Таблица сопряженности')
plt.show()
```
